```python
# importing all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading all data's
df=pd.read_csv("cause_of_deaths.csv")
df
```

```python
# Cleaning the data
df.duplicated().value_counts()
df.drop_duplicates(inplace=True)
df.isnull().sum()
```

```python
# Taking only first 20 values
h=df.head(20)
h
```

# Bar Plot

```python
fig,ax=plt.subplots(2,2,figsize=(15,15),sharey=True)

ax[0,0].bar(h["Year"],h["Meningitis"],color="blue",edgecolor="black",label="Meningitis")
ax[0,0].set_xlabel("Year",fontsize=25,labelpad=20,fontweight='bold')
ax[0,0].set_ylabel("Total Mortality Rate",fontsize=20,fontweight='bold')
ax[0,0].grid(True)
ax[0,0].set_xticks(h["Year"])
ax[0,0].legend(fontsize=16)
ax[0,0].set_xticklabels(h['Year'],rotation=90)

ax[0,1].bar(h["Year"],h["Parkinson's Disease"],color="red",edgecolor="black",label="Parkinson's Disease")
ax[0,1].set_xlabel("Year",fontsize=25,fontweight='bold',labelpad=20)
ax[0,1].set_ylabel("Total Mortality Rate",fontsize=20,fontweight='bold')
ax[0,1].grid(True)
ax[0,1].set_xticks(h["Year"])
ax[0,1].legend(fontsize=16)

ax[0,1].set_xticklabels(h['Year'],rotation=90)

ax[1,0].bar(h["Year"],h["Alzheimer's Disease and Other Dementias"],color="cyan",edgecolor="black",label="Alzheimer's Disease & Other Dementias")
ax[1,0].set_xlabel("Year",fontsize=25,fontweight='bold',labelpad=20)
ax[1,0].set_ylabel("Total Mortality Rate",fontsize=20,fontweight='bold',labelpad=10)
ax[1,0].grid(True)
ax[1,0].set_xticks(h["Year"])
ax[1,0].legend(fontsize=16)

ax[1,0].set_xticklabels(h['Year'],rotation=90)

ax[1,1].bar(h["Year"],h["Malaria"],color="lime",edgecolor="black",label="Malaria")
ax[1,1].set_xlabel("Year",fontsize=25,fontweight='bold',labelpad=20)
ax[1,1].set_ylabel("Total Mortality Rate",fontweight='bold',fontsize=20)
ax[1,1].grid(True)
ax[1,1].set_xticks(h["Year"])
ax[1,1].legend(fontsize=16)

ax[1,1].set_xticklabels(h['Year'],rotation=90)


plt.suptitle("Diseases through different years", fontsize=26,fontweight='bold',y=0.986)
plt.tight_layout()
plt.show()
```

# Scatter Plot

```python
yearly_mortality = h.groupby('Year')['Meningitis'].mean()
lowest_mortality_year = yearly_mortality.idxmax()
lowest_mortality_year
```

```python
fig,ax=plt.subplots()
plt.scatter(h['Year'], h['Malaria'], label='Malaria',color='darkviolet')
plt.scatter(h['Year'], h['Meningitis'], label='Meningitis',color='red')
plt.title('Scatter Plot of Malaria and Meningitis Cases Over Years',fontsize=12,fontweight='bold',y=1.02)
plt.xlabel('Year',fontweight='bold',fontsize=15,labelpad=15)
plt.xticks(h['Year'],rotation=90)
plt.ylabel('Malaria and Meningitis Cases',fontweight='bold',fontsize=15)
plt.legend(loc='upper right',bbox_to_anchor=(1.27,1.02))
plt.grid(True)
plt.show()
```

# Heatmap

```python
# Creating a correlation matrix
selected_columns = ['Meningitis', 'Malaria', "Parkinson's Disease",'Chronic Respiratory Diseases']
selected_data = h[selected_columns]
correlation_matrix = selected_data.corr(method='spearman')
plt.title("Correlation Matrix")
print(correlation_matrix)

# Creating the heatmap using the correlation matrix which we have just created.
sns.heatmap(correlation_matrix,cmap='Blues',annot=True)
plt.title("Heatmap showing correlation between various causes ",pad=10)
plt.show()
```
